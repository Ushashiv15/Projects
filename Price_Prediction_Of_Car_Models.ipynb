```python
d1=d.drop(['lat','lon','ID'],axis=1)
d1
```

```python
d2=d.drop('ID',axis=1)
d2
```

```python
x=d1.drop('price',axis=1)
y=d1['price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=0)
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(x_train,y_train)
```

```python
yp=r.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,yp)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,yp)
```

```python
rs=pd.DataFrame(columns=['price','predicted'])
rs['price']=y_test
rs['predicted']=yp
rs=rs.reset_index()
rs['Id']=rs.index
rs
```

```python
sns.lineplot(x='Id',y='price',data=rs.head(50))
sns.lineplot(x='Id',y='predicted',data=rs.head(50))
```
